```python
def column_cleaning(episode,column_name):
    episode.drop([column_name], axis=1,inplace=True)
    return episode
```

```python
def final_df(episode, name):
    episode = episode.drop([0])
    episode.drop(['index'], axis = 1, inplace = True) 
    episode['episode'] = name
    return episode
```
